# **Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

### What is Anomaly Detection?

**Anomaly Detection**, also known as outlier detection, is the process of identifying rare items, events, or observations that deviate significantly from the majority of the data. These anomalies often indicate critical incidents, such as fraud, system defects, or medical problems. The main goal of anomaly detection is to identify patterns in data that do not conform to expected behavior.

### Types of Anomalies

Anomaly detection typically categorizes anomalies into three main types:

1.  **Point Anomalies (Global Outliers)**
2.  **Contextual Anomalies (Conditional Outliers)**
3.  **Collective Anomalies**

#### 1. Point Anomalies

**Definition:** A point anomaly is a single data instance that is considered anomalous with respect to the rest of the data. It is an individual data point that lies outside the overall distribution or expected range of the dataset, regardless of context.

**Example:**

*   **Credit Card Fraud:** A transaction of $10,000 made by a user who typically spends $50-$100 per purchase. This single transaction is unusual compared to their historical spending patterns.
*   **Sensor Reading:** A temperature sensor in a server room suddenly reports 150째C, while all other readings are around 25-30째C. This single high reading is a point anomaly.

#### 2. Contextual Anomalies

**Definition:** A data instance is considered a contextual anomaly if it deviates significantly from other data points in a specific context. The data point itself might not be anomalous in general, but its occurrence in a particular context makes it unusual. The context could be temporal, spatial, or behavioral.

**Example:**

*   **Website Traffic:** A website typically receives 10,000 visitors per hour during weekdays. On a Saturday morning, it receives 10,000 visitors. While 10,000 visitors per hour is normal for a weekday, it's unusually high for a weekend morning, making it a contextual anomaly.
*   **Body Temperature:** A human body temperature of 37째C is normal. However, if a patient's temperature is 37째C *after* an intensive surgical procedure where a high fever is expected, it might be considered an anomaly in that specific medical context.

#### 3. Collective Anomalies

**Definition:** A collection of related data instances is anomalous with respect to the entire dataset, even if individual data points within the collection are not themselves anomalous. These anomalies occur when a group of data points collectively behave unusually, although each individual point might appear normal when examined in isolation.

**Example:**

*   **Network Intrusion (DDoS Attack):** A distributed denial-of-service (DDoS) attack involves multiple computers sending a small number of requests each to a server. Individually, each request might seem normal, but the collective surge in requests from various sources, targeting the same server simultaneously, constitutes a collective anomaly, indicating a malicious attack.
*   **Human Activity:** A person typically walks in a specific pattern. If a person suddenly starts walking in a perfect circle for an extended period, each step might be normal, but the sequence of steps forms an unusual, collective pattern.

# **Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

### Isolation Forest (iForest)

**Approach:** Isolation Forest is an unsupervised anomaly detection algorithm that works on the principle of isolating anomalies rather than profiling normal data points. It builds an ensemble of Isolation Trees (similar to Decision Trees) on sub-samples of the data. Anomalies are data points that require fewer splits to be isolated from the rest of the data because they are few and different. Normal points, on the other hand, require more splits to be isolated.

**Suitable Use Cases:**
*   High-dimensional datasets.
*   Large datasets where speed is important.
*   Detecting point anomalies (global outliers).
*   Effective when the number of anomalies is small compared to normal instances.

### DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

**Approach:** DBSCAN is a density-based clustering algorithm that groups together data points that are closely packed together, marking as outliers (or 'noise') those points that lie alone in low-density regions. It defines clusters based on a minimum number of points (min_samples) within a certain radius (epsilon or eps) from each core sample. Points that do not belong to any cluster and are not dense enough to form a new cluster are considered anomalies.

**Suitable Use Cases:**
*   Detecting point anomalies in datasets with arbitrarily shaped clusters.
*   When data contains varying densities.
*   Identifying contextual anomalies within dense clusters.
*   Datasets where the number of clusters is unknown.

### Local Outlier Factor (LOF)

**Approach:** Local Outlier Factor (LOF) is an algorithm that computes an anomaly score for each data point based on its local density deviation from its neighbors. A data point is considered an anomaly if its local density is significantly lower than that of its neighbors. This means it's isolated from its surroundings. LOF considers the density around a point and compares it to the density of its neighbors to determine if it's an outlier.

**Suitable Use Cases:**
*   Detecting local anomalies, where a point might be an outlier relative to its local neighborhood but not necessarily to the global dataset.
*   Datasets with varying density distributions.
*   When the nature of anomalies is localized.
*   Smaller to medium-sized datasets, as its computational cost can be higher for very large datasets.

# **Question 3: What are the key components of a Time Series? Explain each with one example.**

### Key Components of a Time Series

A time series is a sequence of data points indexed in time order. Analyzing time series often involves decomposing them into several components that help understand the underlying patterns and make predictions. The primary components are:

1.  **Trend**
2.  **Seasonality**
3.  **Cyclical**
4.  **Irregular / Residual**

#### 1. Trend

**Definition:** The trend component represents the long-term progression or general direction of the time series data. It reflects a persistent, overall increase, decrease, or stagnation over a significant period, smoothing out short-term fluctuations.

**Example:**

*   **Global Temperature:** Over the past century, global average temperatures have shown an **increasing trend**, indicating global warming, despite yearly variations due to natural climate cycles.

#### 2. Seasonality

**Definition:** The seasonality component refers to patterns that repeat over a fixed and known period, such as a day, week, month, or year. These patterns are typically caused by factors like weather changes, holidays, or business practices that occur regularly.

**Example:**

*   **Ice Cream Sales:** Ice cream sales typically show **seasonal patterns**, peaking in summer months (June, July, August) and declining in winter months (December, January, February) each year.

#### 3. Cyclical

**Definition:** The cyclical component describes patterns that occur when data exhibit rises and falls that are not of fixed period, but rather of longer duration, usually more than a year. These fluctuations are often associated with economic business cycles (e.g., recession, expansion) or other multi-year phenomena.

**Example:**

*   **Housing Market Prices:** The housing market often experiences **cyclical patterns** where prices increase for several years (a boom), then stagnate or decrease for a period (a bust), before eventually rising again. These cycles are typically longer than a year and don't have a strict fixed duration.

#### 4. Irregular / Residual

**Definition:** The irregular or residual component represents the random, unpredictable fluctuations in the time series data that cannot be explained by trend, seasonality, or cyclical components. These are often due to unforeseen events or measurement errors.

**Example:**

*   **Stock Prices:** While stock prices might exhibit trends and sometimes seasonal patterns (e.g., year-end rallies), a sudden, sharp drop or rise due to an unexpected news event (like a company scandal or a breakthrough discovery) that doesn't fit any recurring pattern would be considered an **irregular component**.

# **Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

### What is Stationarity in Time Series?

A time series is considered **stationary** if its statistical properties, such as mean, variance, and autocorrelation, are constant over time. This means that the way the series behaves at one point in time is similar to how it behaves at any other point in time. Stationarity is a fundamental assumption for many time series forecasting models, as it simplifies the modeling process.

**Key characteristics of a stationary time series:**

*   **Constant Mean:** The average value of the series does not change over time.
*   **Constant Variance:** The variability of the series remains consistent over time.
*   **Constant Autocorrelation Structure:** The correlation between the series and its lagged versions remains constant over time. This implies that the strength and direction of past values' influence on future values do not change.

### How to Test for Stationarity

There are several ways to test for stationarity:

1.  **Visual Inspection:**
    *   **Plotting the data:** Look for trends (increasing or decreasing patterns), seasonality (repeating patterns), or changing variance (widening or narrowing bands around the mean). A stationary series will typically fluctuate around a constant mean with constant variance.
    *   **Rolling Statistics:** Plot rolling mean and rolling standard deviation. If these statistics remain relatively constant over time, the series is likely stationary.

2.  **Statistical Tests:**
    *   **Augmented Dickey-Fuller (ADF) Test:** This is one of the most common statistical tests for stationarity. It tests the null hypothesis that a unit root is present in a time series (i.e., the series is non-stationary). A low p-value (typically less than 0.05) suggests rejecting the null hypothesis and concluding that the series is stationary.
    *   **Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test:** This test is an alternative to the ADF test. It tests the null hypothesis that a time series is trend-stationary. A high p-value (typically greater than 0.05) suggests failing to reject the null hypothesis, meaning the series is stationary around a trend or mean.

**Example (ADF Test):**

```python
from statsmodels.tsa.stattools import adfuller

def test_stationarity(timeseries):
    # Perform Dickey-Fuller test:
    print('Results of Augmented Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)

    if dfoutput['p-value'] <= 0.05:
        print("The series is stationary (reject H0).")
    else:
        print("The series is non-stationary (fail to reject H0).")
```

### How to Transform a Non-Stationary Series into a Stationary One

If a time series is found to be non-stationary, various transformations can be applied to make it stationary:

1.  **Differencing:**
    *   **Definition:** Differencing involves subtracting the previous observation from the current observation. This helps to remove trends and seasonality.
    *   **First-order differencing:** `diff(t) = original(t) - original(t-1)`
    *   **Seasonal differencing:** `diff(t) = original(t) - original(t-k)` where `k` is the seasonal period.
    *   **Example:** To remove a linear trend, first-order differencing is often effective. To remove seasonality, seasonal differencing is used.

2.  **Transformation (e.g., Logarithmic, Square Root):**
    *   **Definition:** These transformations can help stabilize the variance of a time series, especially when the variance increases with the mean. For instance, taking the logarithm of the series can reduce the magnitude of fluctuations and make the variance more constant.
    *   **Example:** If the amplitude of seasonal fluctuations increases with the level of the series, a logarithmic transformation can often make the seasonality more additive and the variance more stable.

3.  **Decomposition:**
    *   **Definition:** Decomposing a time series separates it into its components: trend, seasonality, and residual. Once decomposed, the non-stationary components (trend and seasonality) can be removed, leaving a stationary residual component.
    *   **Example:** Using statistical methods like `seasonal_decompose` from `statsmodels` to extract and then remove the trend and seasonal components.

# **Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**


### Differentiating Time Series Models

Time series forecasting relies on various models that capture different aspects of data dependencies over time. Here, we'll differentiate between some common models: AR, MA, ARIMA, SARIMA, and SARIMAX.

#### 1. Autoregressive (AR) Model

**Structure:** An AR(p) model predicts the future value based on a linear combination of its past 'p' values. It assumes that the current value is a weighted sum of previous values plus a white noise error term.

*   **Equation:** $Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + ... + \phi_p Y_{t-p} + \epsilon_t$
    *   $Y_t$: Current value
    *   $Y_{t-i}$: Past values (lagged values)
    *   $\phi_i$: Autoregressive coefficients
    *   $c$: Constant
    *   $\epsilon_t$: White noise error term

**Application:** Suitable for time series that show a correlation with their own past values, exhibiting memory. Often used when the primary characteristic is a decaying autocorrelation function (ACF).

#### 2. Moving Average (MA) Model

**Structure:** An MA(q) model predicts the future value based on a linear combination of past 'q' white noise error terms. It models the dependency between an observation and a residual error from a moving average process applied to past observations.

*   **Equation:** $Y_t = c + \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + ... + \theta_q \epsilon_{t-q}$
    *   $Y_t$: Current value
    *   $\epsilon_t$: Current white noise error term
    *   $\epsilon_{t-i}$: Past white noise error terms (lagged errors)
    *   $\theta_i$: Moving average coefficients
    *   $c$: Constant

**Application:** Effective for time series where the current value depends on past forecast errors. Often used when the primary characteristic is a decaying partial autocorrelation function (PACF).

#### 3. Autoregressive Integrated Moving Average (ARIMA) Model

**Structure:** An ARIMA(p, d, q) model combines AR, I (Integrated), and MA components. It's used for non-stationary time series, where 'd' represents the number of differencing operations required to make the series stationary.

*   **Equation:** The ARIMA model applies differencing 'd' times to the time series to make it stationary, and then applies an AR(p) and MA(q) model to the differenced series. If $Y'_t$ is the differenced series, then it follows an ARMA(p, q) process:
    $Y'_t = c + \phi_1 Y'_{t-1} + ... + \phi_p Y'_{t-p} + \epsilon_t + \theta_1 \epsilon_{t-1} + ... + \theta_q \epsilon_{t-q}$

**Application:** Widely used for forecasting non-seasonal time series data. It can capture trends by differencing and model autocorrelation with AR and MA terms. Requires the time series to be made stationary (or differenced to be stationary) before modeling.

#### 4. Seasonal Autoregressive Integrated Moving Average (SARIMA) Model

**Structure:** A SARIMA(p, d, q)(P, D, Q)m model extends ARIMA by adding seasonal components. It handles time series with both non-seasonal and seasonal patterns.

*   **Parameters:**
    *   **(p, d, q):** Non-seasonal AR, Integrated, and MA orders.
    *   **(P, D, Q)m:** Seasonal AR, Integrated, and MA orders, where 'm' is the number of observations per season (e.g., 12 for monthly data, 4 for quarterly data).

**Application:** Ideal for time series exhibiting clear seasonal patterns in addition to non-seasonal trends and correlations. Examples include monthly sales data, daily electricity consumption, etc.

#### 5. Seasonal Autoregressive Integrated Moving Average with Exogenous Regressors (SARIMAX) Model

**Structure:** A SARIMAX(p, d, q)(P, D, Q)m model is an extension of SARIMA that includes exogenous (external) variables. These are variables that are not part of the time series itself but can influence its behavior.

*   **Parameters:** Same as SARIMA, plus additional exogenous variables ($X_t$).
*   **Equation (Conceptual):** SARIMA(p, d, q)(P, D, Q)m + $\beta_1 X_{1,t} + \beta_2 X_{2,t} + ... + \epsilon_t$

**Application:** Used when external factors significantly impact the time series. For example, forecasting sales might involve exogenous variables like marketing spend, competitor actions, or holidays. It allows for a more comprehensive model that incorporates external influences.

## Anomaly Detection for Streaming Energy Data

### Choice of Anomaly Detection Algorithm: Isolation Forest

#### 1. Characteristics of Real-time, Streaming Energy Consumption Data

Real-time energy consumption data typically presents the following characteristics:

*   **High Volume and Velocity**: Data arrives continuously and rapidly (every 15 minutes), leading to a large volume of time series data.
*   **Seasonality and Trends**: Energy usage patterns often exhibit strong daily, weekly, and annual seasonality, along with long-term trends influenced by population growth, economic factors, or energy efficiency initiatives.
*   **Multi-dimensional**: Besides energy usage, features like `timestamp`, `region`, `weather conditions` (temperature, humidity), and potentially `day_of_week`, `hour_of_day`, `holiday_status` add to the dimensionality.
*   **Dynamic Nature**: Underlying patterns can shift due to external events (e.g., new industrial plants, policy changes, climate shifts).
*   **Local and Global Anomalies**: Spikes or drops can be globally anomalous (e.g., a massive grid failure) or locally anomalous (e.g., one region showing unusual usage for its specific context).
*   **Non-stationary**: Due to trends and seasonality, the statistical properties of the series change over time, making it non-stationary.

#### 2. Evaluation of Algorithms for Streaming Data

Let's evaluate Isolation Forest, LOF, and DBSCAN based on the above characteristics:

*   **Isolation Forest (iForest)**:
    *   **Strengths**: Highly efficient for high-dimensional and large datasets. It identifies anomalies by isolating them with fewer splits in a tree structure, which is computationally faster than distance-based methods. It doesn't assume data distribution and is effective for global anomalies. It naturally handles numerical data.
    *   **Weaknesses**: Can struggle with very localized anomalies that are dense relative to their immediate neighbors but sparse globally. Interpretability can be moderate (path length is the score).
    *   **Streaming Suitability**: Its tree-based nature allows for online updates (though not natively incremental in `sklearn`), and its speed is crucial for high-velocity data.

*   **Local Outlier Factor (LOF)**:
    *   **Strengths**: Excellent at detecting local anomalies by comparing a point's density to that of its neighbors. Provides an anomaly score indicating the degree of outlierness.
    *   **Weaknesses**: Computationally expensive, especially for high-dimensional data, as it relies on distance calculations between all points and their neighbors. `n_neighbors` parameter choice is crucial. Can be slow for large streaming datasets.
    *   **Streaming Suitability**: Less suitable for true real-time streaming due to its computational cost and the need to re-evaluate local densities as new data arrives.

*   **DBSCAN**:
    *   **Strengths**: Good for detecting point anomalies (noise) in datasets with varying densities and arbitrarily shaped clusters. No need to specify the number of clusters.
    *   **Weaknesses**: Highly sensitive to `eps` (radius) and `min_samples` parameters. Can struggle with very high-dimensional data without feature engineering. Not designed for anomaly scoring, rather for identifying noise points as a byproduct of clustering.
    *   **Streaming Suitability**: Not inherently designed for streaming data. Re-running DBSCAN on a continually expanding dataset or sliding window can be computationally prohibitive.

#### 3. Chosen Algorithm: Isolation Forest

For real-time streaming energy consumption data, **Isolation Forest** is the most appropriate choice.

**Justification:**

1.  **Computational Efficiency for Streaming**: Its tree-based approach means it does not rely on distance metrics across the entire dataset for every new data point, making it significantly faster than LOF or DBSCAN. This is critical for processing data arriving every 15 minutes.
2.  **Scalability**: It handles high-dimensional data effectively, which is important given the multiple features (timestamp-derived features, region, weather).
3.  **Effectiveness for Global Anomalies**: While energy data can have local anomalies, significant spikes or drops in overall energy usage or within a region, indicative of critical incidents (equipment failure, cyber-attack, sudden demand surge), are often global outliers compared to the normal range of values. iForest excels at identifying these.
4.  **No Density Assumptions**: It doesn't assume a specific distribution or density, which is robust for complex energy consumption patterns.

#### 4. Real-time Implementation of Isolation Forest in a Streaming Context

Implementing Isolation Forest for streaming data requires a strategy to handle continuous data flow and maintain model relevance.

1.  **Data Preprocessing and Feature Engineering**: Each incoming 15-minute data point will be enriched with relevant features:
    *   **Time-based**: `hour_of_day`, `day_of_week`, `day_of_month`, `month_of_year`, `is_weekend`, `is_holiday`.
    *   **Lagged Features**: Previous `energy usage` values (e.g., usage 15 min ago, 1 hour ago, 24 hours ago, 1 week ago) to capture temporal dependencies.
    *   **External Features**: Current `weather conditions` (temperature, humidity).
    *   **Region-specific**: Encode `region` if not already numerical.
    All these features form the input vector for the Isolation Forest.

2.  **Sliding Window Approach**: Instead of retraining on all historical data constantly, a sliding window approach is used.
    *   A fixed-size window (e.g., data from the last 24 hours, 7 days, or longer depending on seasonality) is maintained. As new data arrives, the oldest data leaves the window.
    *   The Isolation Forest model is trained on the data within this sliding window.

3.  **Model Retraining Strategy**: Full retraining of iForest on every new data point is too resource-intensive. Options include:
    *   **Periodic Retraining**: Retrain the Isolation Forest model at regular intervals (e.g., hourly, daily, or weekly) on the current sliding window of data. This captures evolving patterns.
    *   **Ensemble of Models**: Maintain an ensemble of Isolation Forest models, where new models are trained on recent data chunks and combined with older models, allowing for adaptive learning.
    *   **Incremental Learning (if a suitable library is used)**: While `sklearn`'s Isolation Forest is not natively incremental, specialized libraries or custom implementations could support updating the model with new data without full retraining, though this is more complex.

5.  **Threshold Management**: The anomaly threshold can be static or dynamically adjusted. Dynamic adjustment could involve:

6.  **Alerting and Action**: When an anomaly is detected, an alert is sent to operations teams, potentially including the `timestamp`, `region`, `usage deviation`, and `score`. This allows for immediate investigation and intervention (e.g., checking equipment, rerouting power).

2.  **SARIMA (Seasonal Autoregressive Integrated Moving Average):** An extension of ARIMA that explicitly models seasonal patterns. However, it *cannot directly incorporate exogenous variables*.
3.  **SARIMAX (Seasonal Autoregressive Integrated Moving Average with Exogenous Regressors):** Extends SARIMA by allowing the inclusion of exogenous variables. ### Justification for Model Choice
Given the characteristics of the energy demand data and the need for short-term forecasting, **SARIMAX** is the most appropriate model.

*   **Seasonality:** Instead of typical yearly seasonality (m=12 for monthly data), we will encounter much shorter seasonal cycles:
    *   **Daily Seasonality:** 24 hours * 4 (15-minute intervals/hour) = 96 observations per day. So, `m=96` for daily seasonality.
    *   **Weekly Seasonality:** 7 days * 96 intervals/day = 672 observations per week. For weekly patterns, `m=672`.

*   **Computational Considerations:** With 96 observations per day, a 15-minute dataset will grow very rapidly. Fitting SARIMAX models with very high seasonal orders (`m=96` or `m=672`) can be computationally intensive. This might necessitate:
    *   **Data Aggregation:** For longer-term forecasts or less critical periods, aggregating data to hourly or daily resolution might be considered.
    *   **Parallel Processing:** Utilizing parallel computing for model fitting.
    *   **Approximation Methods:** Employing approximate SARIMAX methods if available.
    *   **Rolling Forecasts:** Implementing rolling forecasts to retrain the model frequently on recent data, ensuring adaptability to changing patterns.
*   **Lag Selection:** The `p` and `q` orders for AR and MA components might also be larger to capture finer-grained dependencies.

*   **Other Exogenous Variables:** Features like day of the week, hour of the day (already captured by seasonality but can be reinforced), public holidays, and special events (e.g., major sporting events) can also be one-hot encoded and included as exogenous variables.

## Explain Validation and Monitoring Strategy

### Comprehensive Strategy for Model Validation and Monitoring

To ensure the reliability and effectiveness of both the energy demand forecasting and anomaly detection models in a power grid monitoring company, a robust validation and continuous monitoring strategy is essential. This strategy encompasses appropriate metrics, mechanisms for detecting data and model drift, and intelligent re-training approaches.

### 1. Validation Metrics for Forecasting Model (Energy Demand)

For short-term energy demand forecasting, accuracy metrics are crucial to understand how well the model predicts future consumption. Given that energy usage is a continuous numerical variable, regression-based metrics are suitable:

*   **Mean Absolute Error (MAE):**
    *   **Relevance:** MAE measures the average magnitude of the errors in a set of forecasts, without considering their direction. It's robust to outliers and is expressed in the same units as the target variable (e.g., kWh), making it easily interpretable by operations teams regarding the average prediction miss.

*   **Root Mean Squared Error (RMSE):**
    *   **Relevance:** RMSE is the square root of the average of the squared errors. It penalizes larger errors more severely than MAE, which is critical in energy forecasting where large deviations (e.g., under-predicting peak demand) can have significant operational and financial consequences. Also in kWh, but less intuitive for average error.

*   **Mean Absolute Percentage Error (MAPE):**
    *   **Relevance:** MAPE expresses the accuracy as a percentage of the actual value. This is highly useful for comparing forecast performance across different regions or time periods with varying scales of energy demand. A MAPE of 5% is easily understood across all stakeholders.

*   **R-squared (Coefficient of Determination):**
    *   **Relevance:** R-squared indicates the proportion of the variance in the dependent variable that is predictable from the independent variables. It provides a measure of how well future samples are likely to be predicted by the model, giving a general sense of explanatory power.

### 2. Validation Metrics for Anomaly Detection Model

Anomaly detection models, especially for real-time spikes or drops, are often evaluated using classification metrics, provided there is some form of labeled anomalous data or expert feedback:

*   **Precision:**

    *   **Anomaly Detection Metrics**: Precision, Recall, F1-score, False Positive Rate (FPR), and False Negative Rate (FNR) were chosen to balance avoiding alert fatigue with detecting critical events.
    *   **Continuous Monitoring**: Data drift (changes in input data distributions via statistical tests or feature monitoring) and model drift (deterioration of performance via error monitoring or anomaly count monitoring) will be continuously tracked.
    *   **Retraining and Adaptation**: Models will undergo scheduled retraining or triggered retraining (based on performance degradation, detected drift, or significant external events). New model versions will be deployed safely using A/B testing methods like shadow mode or canary deployments.
