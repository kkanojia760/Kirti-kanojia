# **1-What is hypothesis testing in statistics?**



Hypothesis testing is a statistical method used to determine if there is enough evidence in a sample data to determine that a certain condition or statement about the population is true. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), collecting data, and using statistical tests (like chisquare, t-test etc.) to evaluate the evidence against the null hypothesis. The goal is to decide whether to reject or fail to reject the null hypothesis based on the p-value and a predetermined significance level (alpha).

# **What is the null hypothesis, and how does it differ from the alternative hypothesis?**


In hypothesis testing, we have two competing statements about a population:

*   **Null Hypothesis (H0):** This is the statement of no effect, no difference, or no relationship. It's often the status quo or the commonly accepted idea. We assume the null hypothesis is true until we have enough evidence to suggest otherwise. Think of it as the "innocent until proven guilty" in statistics.

*   **Alternative Hypothesis (H1 or Ha):** This is the statement that contradicts the null hypothesis. It's what you are trying to find evidence for. It suggests there is an effect, a difference, or a relationship.

The goal of hypothesis testing is to determine if there is enough statistical evidence to reject the null hypothesis in favor of the alternative hypothesis.

# **Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.**

The **significance level (alpha or α)** is a threshold you set *before* you start your hypothesis test. It represents the maximum probability of rejecting the null hypothesis when it is actually true (this is also known as a Type I error or a false positive).

Here's its role in deciding the outcome:

1.  **You compare the p-value to the significance level:** After running your statistical test, you get a p-value. The p-value is the probability of observing your data (or something more extreme) if the null hypothesis were true.
2.  **Decision Rule:**
    *   If the **p-value is less than or equal to alpha (p ≤ α)**, you **reject the null hypothesis**. This means your data is statistically significant at the chosen alpha level, suggesting there's enough evidence to support the alternative hypothesis.
    *   If the **p-value is greater than alpha (p > α)**, you **fail to reject the null hypothesis**. This means your data is not statistically significant at the chosen alpha level, and you don't have enough evidence to support the alternative hypothesis.

Commonly used significance levels are 0.05 (5%), 0.01 (1%), and 0.10 (10%). Choosing a smaller alpha makes it harder to reject the null hypothesis, reducing the risk of a Type I error but increasing the risk of a Type II error (failing to reject the null hypothesis when the alternative hypothesis is true).

# **What are Type I and Type II errors? Give examples of each.**

In hypothesis testing, there are two main types of errors we can make:

*   **Type I Error (False Positive):** This occurs when you **reject the null hypothesis** when it is actually **true**. You conclude there is an effect or difference when there isn't one. The probability of making a Type I error is equal to your significance level (alpha, α).

    *   **Example:** A medical test for a rare disease incorrectly says a healthy person has the disease. The null hypothesis is that the person is healthy (no disease). The test rejects the null hypothesis, but it's wrong.

*   **Type II Error (False Negative):** This occurs when you **fail to reject the null hypothesis** when it is actually **false**. You fail to detect an effect or difference that actually exists. The probability of making a Type II error is denoted by beta (β).

    *   **Example:** A medical test for a disease incorrectly says a sick person is healthy. The null hypothesis is that the person is healthy. The test fails to reject the null hypothesis, but it's wrong because the person is sick.

# **What is the difference between a Z-test and a T-test? Explain when to use each.**

Both the Z-test and the T-test are statistical tests used to compare means. The main difference lies in the information you have about the population standard deviation:

*   **Z-test:** Used when you **know the population standard deviation (σ)**. It's also used when the sample size is large (typically n > 30), even if the population standard deviation is unknown, because the sample standard deviation becomes a good estimate of the population standard deviation. The test statistic follows a standard normal distribution (Z-distribution).

    *   **When to use:** Comparing a sample mean to a known population mean with a known population standard deviation, or with a large sample size.

*   **T-test:** Used when you **do not know the population standard deviation (σ)** and have to estimate it using the sample standard deviation (s). The test statistic follows a t-distribution, which has heavier tails than the Z-distribution, accounting for the extra uncertainty introduced by estimating the population standard deviation. The shape of the t-distribution depends on the sample size (specifically, the degrees of freedom, which is related to the sample size). As the sample size increases, the t-distribution approaches the Z-distribution.

    *   **When to use:** Comparing a sample mean to a known population mean when the population standard deviation is unknown and the sample size is small to moderate. Comparing the means of two groups when the population standard deviations are unknown.

In summary: Use a Z-test if you know the population standard deviation or have a large sample size. Use a T-test if you don't know the population standard deviation and have a small to moderate sample size.

### What Z-scores Represent

A Z-score (or standard score) tells you how many standard deviations away from the mean an individual data point is.

*   **A Z-score of 0** means the data point is exactly at the mean.
*   **A positive Z-score** indicates the data point is above the mean. For example, a Z-score of +1 means the data point is one standard deviation above the mean.
*   **A negative Z-score** indicates the data point is below the mean. For example, a Z-score of -2 means the data point is two standard deviations below the mean.

Z-scores are useful for:

*   **Standardizing data:** Allowing you to compare data points from different distributions with different means and standard deviations.
*   **Identifying outliers:** Data points with very high or very low Z-scores (e.g., typically beyond +3 or -3) may be considered outliers.
*   **Calculating probabilities:** In a normal distribution, Z-scores can be used with Z-tables or statistical software to find the probability of observing a value less than, greater than, or between specific values.
