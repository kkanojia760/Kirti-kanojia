# **1. What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

### Descriptive Statistics

**Descriptive statistics** are used to summarize and describe the characteristics of a dataset. They provide a concise overview of the data, helping us understand its main features. Descriptive statistics do not involve making inferences or generalizations about a larger population; they only describe the data that has been collected.

**Common measures include:**

*   **Measures of Central Tendency:**
    *   **Mean:** The average of all values.
    *   **Median:** The middle value when the data is ordered.
    *   **Mode:** The most frequently occurring value.

*   **Measures of Variability (Spread):**
    *   **Range:** The difference between the highest and lowest values.
    *   **Variance:** The average of the squared differences from the mean.
    *   **Standard Deviation:** The square root of the variance, indicating the typical distance of data points from the mean.

*   **Frequency Distributions:** Tables or graphs that show how often each value or range of values occurs.

**Example:**

Imagine you collect the test scores of 30 students in a class:

`[75, 80, 85, 70, 90, 78, 92, 88, 76, 83, 95, 68, 81, 89, 77, 84, 91, 72, 86, 79, 93, 65, 87, 82, 74, 94, 71, 80, 85, 73]`

Using descriptive statistics, you might calculate:

*   **Mean Score:** 82.07
*   **Median Score:** 82.5
*   **Mode:** 80 and 85 (bimodal)
*   **Range:** 95 - 65 = 30
*   **Standard Deviation:** Approximately 7.97

These statistics describe *only* the performance of these specific 30 students in this particular test. They don't tell us anything about how students in other classes or schools performed.

### Inferential Statistics

**Inferential statistics** go beyond simply describing data. They involve making predictions, inferences, and generalizations about a larger population based on a sample of data drawn from that population. This is crucial when it's impractical or impossible to study an entire population.

**Key concepts include:**

*   **Hypothesis Testing:** A formal procedure to determine if there is enough evidence in a sample to reject a null hypothesis about the population.
*   **Confidence Intervals:** A range of values within which a population parameter is estimated to lie, with a certain level of confidence.
*   **Regression Analysis:** Used to model the relationship between a dependent variable and one or more independent variables.
*   **ANOVA (Analysis of Variance):** Used to compare means of three or more groups.

**Example:**

Continuing with the student test scores, let's say the 30 students were a *random sample* from a much larger school with 500 students.

Using inferential statistics, you might:

1.  **Estimate the average test score for all 500 students in the school.** Based on the sample mean of 82.07, you could construct a 95% confidence interval, perhaps stating that "we are 95% confident that the true average test score for all students in the school is between 79.2 and 84.9."
2.  **Test a hypothesis:** Suppose the principal claims that the average test score for all students in the school is 85. You could use a hypothesis test to determine if your sample data provides enough evidence to reject the principal's claim.
3.  **Compare teaching methods:** If 15 students were taught using method A and 15 using method B, inferential statistics could be used to determine if there's a statistically significant difference in test scores between the two teaching methods for the *entire school population*.

**In summary:**

*   **Descriptive statistics** summarize the characteristics of *known* data.
*   **Inferential statistics** use *sample* data to make educated guesses and draw conclusions about a *larger population*.

# **2. What is sampling in statistics? Explain the differences between random and stratified sampling.**

**Sampling** in statistics is the process of selecting a subset of individuals or items from a larger population to make inferences about the entire population. It is often impractical or impossible to collect data from every member of a population, so researchers use sampling to gather information from a representative group. The goal of sampling is to obtain a sample that accurately reflects the characteristics of the population, thereby minimizing bias and ensuring that the conclusions drawn from the sample can be generalized to the population.

### Differences Between Random and Stratified Sampling

#### 1. Random Sampling (Simple Random Sampling)

**Definition:** Simple random sampling is a method where every member of the population has an equal and independent chance of being selected for the sample. This means that each possible sample of a given size has an equal probability of being chosen.

**How it works:**
*   Assign a unique number to each member of the population.
*   Use a random number generator or a lottery method to select the required number of individuals.

**Characteristics:**
*   **Unbiased:** If done correctly, it minimizes bias, as every unit has an equal chance.
*   **Representativeness:** Assumes that with a sufficiently large sample size, the sample will naturally represent the population's diversity.
*   **Simplicity:** Conceptually straightforward to understand and implement for homogeneous populations.

**Advantages:**
*   Easy to understand and implement.
*   High external validity (generalizability) if the sample is truly random and large enough.

**Disadvantages:**
*   May not be truly representative if the population is heterogeneous and the sample size is small, potentially missing important subgroups.
*   Can be challenging to obtain a complete list of the entire population (sampling frame).
*   Less efficient if the population has distinct subgroups that need to be represented.

**Example:**
Imagine you want to survey 100 students from a university with 10,000 students. You assign each student a number from 1 to 10,000 and then use a random number generator to select 100 unique numbers. The students corresponding to those numbers form your simple random sample.

#### 2. Stratified Sampling

**Definition:** Stratified sampling is a method that involves dividing the population into distinct subgroups, called **strata**, based on shared characteristics (e.g., age, gender, income level, education). After forming these strata, a simple random sample is then drawn from *each* stratum.

**How it works:**
*   Divide the population into mutually exclusive and collectively exhaustive strata.
*   Perform simple random sampling within each stratum.
*   The sample size from each stratum can be proportional to the stratum's size in the population (proportional stratified sampling) or allocated based on other criteria like variability within the stratum (disproportionate stratified sampling).

**Characteristics:**
*   **Ensures Representation:** Guarantees that specific subgroups are adequately represented in the sample, which is particularly useful for heterogeneous populations.
*   **Reduces Sampling Error:** Can lead to more precise estimates than simple random sampling, especially when there's significant variability between strata.
*   **Complex:** Requires prior knowledge of the population's characteristics to form effective strata.

**Advantages:**
*   Provides more precise estimates and reduces sampling error for heterogeneous populations.
*   Ensures representation of important subgroups, allowing for subgroup analysis.
*   Can be more statistically efficient than simple random sampling.

**Disadvantages:**
*   Requires comprehensive knowledge of the population's characteristics to define strata.
*   More complex to implement than simple random sampling.
*   If strata are not well-defined, it can introduce bias.

**Example:**
Continuing with the university example, suppose you want to ensure that each academic year (freshman, sophomore, junior, senior) is proportionally represented in your sample of 100 students. You would divide the 10,000 students into four strata based on their academic year. If freshmen make up 30% of the university, you would randomly select 30 freshmen. If sophomores are 25%, you would randomly select 25 sophomores, and so on, until you have 100 students.

### Key Differences Summarized:

| Feature              | Random Sampling                                       | Stratified Sampling                                     |
| :------------------- | :---------------------------------------------------- | :------------------------------------------------------ |
| **Population Division** | No division; treated as one homogeneous group         | Divided into distinct, non-overlapping subgroups (strata) |
| **Selection Method** | Random selection from the entire population           | Random selection from *each* stratum                  |
| **Representativeness** | Assumed for large samples; not guaranteed for small, heterogeneous populations | Guaranteed for each defined subgroup                      |
| **Bias Reduction**   | Minimizes overall bias                                | Minimizes bias within subgroups; better for heterogeneous populations |
| **Complexity**       | Simpler to implement                                  | More complex; requires prior population knowledge         |
| **Precision**        | Lower precision for heterogeneous populations         | Higher precision, especially for heterogeneous populations |
| **Use Case**         | Homogeneous populations; quick and easy studies       | Heterogeneous populations where subgroup analysis is important |

# **Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

### Mean, Median, and Mode: Measures of Central Tendency

Measures of central tendency are statistical values that represent the center or typical value of a dataset. They provide a concise summary of the data, helping us understand where the majority of the values lie.

#### 1. Mean

**Definition:** The **mean** (also known as the arithmetic average) is calculated by summing all the values in a dataset and then dividing by the total number of values. It is the most commonly used measure of central tendency.

**Formula:**
$\text{Mean (}\bar{x}\text{)} = \frac{\sum_{i=1}^{n} x_i}{n}$
Where:
*   $\sum x_i$ is the sum of all values in the dataset
*   $n$ is the number of values in the dataset

**Example:** For the dataset `[10, 12, 15, 18, 20]`
Mean = $(10 + 12 + 15 + 18 + 20) / 5 = 75 / 5 = 15$

**Why it's important:**
*   **Balances out values:** The mean takes into account every value in the dataset, providing a measure that balances out all the data points.
*   **Widely understood:** It's a familiar concept and easy to calculate and interpret.
*   **Foundation for other statistics:** Many advanced statistical analyses and models are built upon the mean.
*   **Useful for symmetric distributions:** It works best when the data is symmetrically distributed without extreme outliers.

**Limitation:** The mean is highly sensitive to outliers (extreme values). A single very large or very small value can significantly skew the mean, making it less representative of the typical value in such cases.

#### 2. Median

**Definition:** The **median** is the middle value in a dataset when the values are arranged in ascending or descending order. If the dataset has an odd number of values, the median is the single middle value. If it has an even number of values, the median is the average of the two middle values.

**Example:**
*   For an odd number of values: Dataset `[10, 12, 15, 18, 20]`. Sorted: `[10, 12, **15**, 18, 20]`. Median = 15.
*   For an even number of values: Dataset `[10, 12, 15, 18, 20, 22]`. Sorted: `[10, 12, **15, 18**, 20, 22]`. Median = $(15 + 18) / 2 = 16.5$.

**Why it's important:**
*   **Resistant to outliers:** Unlike the mean, the median is not affected by extreme outliers, making it a more robust measure of central tendency for skewed distributions.
*   **Represents the 'typical' value:** It literally represents the middle point of the data, where 50% of the values are below it and 50% are above it.
*   **Useful for ordinal data:** It can be used for data that has a meaningful order but where the differences between values might not be consistent (e.g., satisfaction ratings: very unhappy, unhappy, neutral, happy, very happy).

**Limitation:** It does not consider all values in the dataset, only their position, and therefore might not fully reflect the overall distribution.

#### 3. Mode

**Definition:** The **mode** is the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), multiple modes (multimodal), or no mode if all values appear with the same frequency.

**Example:**
*   Dataset `[10, 12, 15, 12, 20, 12, 18]`. Mode = 12 (appears 3 times).
*   Dataset `[10, 12, 15, 12, 20, 15, 18]`. Modes = 12 and 15 (bimodal).
*   Dataset `[10, 12, 15, 18, 20]`. No mode (each value appears once).

**Why it's important:**
*   **Applicable to all data types:** It is the only measure of central tendency that can be used with nominal (categorical) data, where values are categories without a natural order (e.g., favorite color, type of car).
*   **Identifies common occurrences:** It highlights the most popular or frequently occurring observation, which can be valuable in many contexts (e.g., most popular product, most common age group).
*   **Easy to understand:** It's intuitive to grasp which value occurs most often.

**Limitation:** The mode may not be unique, and for continuous data, it might not exist or be very meaningful if no single value repeats exactly.

### Summary of Importance:

These three measures of central tendency are important because they:

1.  **Provide a quick summary:** They give a single number that attempts to describe the whole dataset, making complex data more digestible.
2.  **Help in understanding typical values:** They allow us to grasp what a 'normal' or 'average' observation looks like within the data.
3.  **Reveal data characteristics:** The relationship between the mean, median, and mode can indicate the skewness of the data's distribution (e.g., if mean > median, the data is likely right-skewed).
4.  **Aid in decision-making:** By understanding the central tendency, researchers, businesses, and policymakers can make informed decisions based on what is typical or most frequent in a population or sample.

# **Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**


### Skewness and Kurtosis

**Skewness** and **Kurtosis** are two important statistical measures that help us understand the shape and characteristics of a dataset's distribution, beyond just its central tendency (mean, median, mode) and variability (range, standard deviation).

#### 1. Skewness

**Definition:** Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it tells us if the data is concentrated more on one side of the mean than the other, resulting in a 'tail' on one side.

*   **Zero Skewness:** If a distribution is perfectly symmetrical (like a normal distribution), its skewness is zero. The mean, median, and mode are all approximately equal.
*   **Positive Skewness (Right-Skewed):** When a distribution has a long tail extending to the right. This means the majority of the data is concentrated on the left side, and there are some relatively large values that pull the mean to the right of the median.
*   **Negative Skewness (Left-Skewed):** When a distribution has a long tail extending to the left. This means the majority of the data is concentrated on the right side, and there are some relatively small values that pull the mean to the left of the median.

**What does a positive skew imply about the data?**

A **positive skew (or right-skewed distribution)** implies the following:

1.  **Tail on the Right:** The right side of the distribution is longer or fatter, indicating a few extremely high values in the dataset.
2.  **Order of Central Tendencies:** In a positively skewed distribution, the **mode** is typically less than the **median**, which is less than the **mean** (Mode < Median < Mean). The mean is pulled towards the higher values by the outliers on the right.
3.  **Data Concentration:** Most of the data points are clustered towards the lower values, with a few observations having significantly higher values.
4.  **Real-world Examples:** Examples often include income distribution (most people earn less, a few earn a lot), housing prices, or test scores with a floor effect (most students score high, a few score very low, making the 'low' tail longer).

**Visual Representation:** Imagine a histogram where the peak is on the left, and the bars gradually get shorter as you move towards the right, extending further to the right than to the left.

#### 2. Kurtosis

**Definition:** Kurtosis is a measure of the 'tailedness' of the probability distribution of a real-valued random variable. It describes the shape of the tails and the peakedness of a distribution relative to a normal distribution. It tells us how many outliers are present.

*   **Mesokurtic (Normal Kurtosis):** A distribution with kurtosis similar to that of a normal distribution. It's often used as a baseline, with a kurtosis value of 3 (or 0 if using excess kurtosis).
*   **Leptokurtic (High Kurtosis):** A distribution with fatter tails and a sharper peak than a normal distribution. This indicates that the data has more extreme outliers than a normal distribution would, meaning more values are concentrated around the mean and in the tails.
*   **Platykurtic (Low Kurtosis):** A distribution with thinner tails and a flatter peak than a normal distribution. This suggests that the data has fewer extreme outliers than a normal distribution, meaning values are more spread out from the mean, and the peak is less pronounced.

**Importance:**

*   **Skewness** helps us understand the direction and extent of the asymmetry of a distribution, which is crucial for choosing appropriate statistical tests or models. For instance, many statistical methods assume normally distributed data (zero skewness).
*   **Kurtosis** informs us about the presence of outliers and the spread of values in the tails. High kurtosis implies a greater risk of extreme outcomes, which is particularly important in fields like finance for risk assessment. It also helps to understand if the data is concentrated or spread out more than a normal distribution.

# **Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers. numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28] (Include your Python code and output in the code box below.)**

```python
import collections

def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 1:
        return sorted_numbers[n // 2]
    else:
        mid1 = sorted_numbers[n // 2 - 1]
        mid2 = sorted_numbers[n // 2]
        return (mid1 + mid2) / 2

def calculate_mode(numbers):
    counts = collections.Counter(numbers)
    max_count = max(counts.values())
    modes = [key for key, value in counts.items() if value == max_count]
    return sorted(modes)

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_value = calculate_mean(numbers)
median_value = calculate_median(numbers)
mode_values = calculate_mode(numbers)

print(f"Given list of numbers: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_values}")
```

# **Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] (Include your Python code and output in the code box below.)**

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to numpy arrays for easier calculation
array_x = np.array(list_x)
array_y = np.array(list_y)

# Compute Covariance
# np.cov returns a covariance matrix. For two arrays, it returns a 2x2 matrix:
# [[cov(x,x), cov(x,y)],
#  [cov(y,x), cov(y,y)]]
# We are interested in cov(x,y) which is at index [0, 1] or [1, 0].
covariance_matrix = np.cov(array_x, array_y)
covariance = covariance_matrix[0, 1]

# Compute Correlation Coefficient
# np.corrcoef also returns a 2x2 correlation matrix.
# The correlation coefficient between x and y is at index [0, 1] or [1, 0].
correlation_matrix = np.corrcoef(array_x, array_y)
correlation_coefficient = correlation_matrix[0, 1]

print(f"Dataset X: {list_x}")
print(f"Dataset Y: {list_y}")
print(f"Covariance between X and Y: {covariance}")
print(f"Correlation Coefficient between X and Y: {correlation_coefficient}")
```

# **Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] (Include your Python code and output in the code box below.)**

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create a figure and an axes object
fig, ax = plt.subplots(figsize=(8, 6))

# Create the boxplot
ax.boxplot(data, vert=False) # vert=False for horizontal boxplot

# Add labels and title
ax.set_title('Boxplot of Numeric Data')
ax.set_xlabel('Value')
ax.set_yticks([]) # Hide y-axis ticks as there's only one boxplot

# Show the plot
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

# Identify outliers programmatically using IQR method
# Calculate Q1 (25th percentile)
Q1 = np.percentile(data, 25)

# Calculate Q3 (75th percentile)
Q3 = np.percentile(data, 75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Data: {data}")
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")
```

### Explanation of the Boxplot and Outliers

The boxplot visually summarizes the distribution of the data based on five key numbers:

1.  **Minimum:** The smallest value in the dataset (excluding outliers).
2.  **First Quartile (Q1):** The 25th percentile, meaning 25% of the data falls below this value.
3.  **Median (Q2):** The 50th percentile, which is the middle value of the dataset.
4.  **Third Quartile (Q3):** The 75th percentile, meaning 75% of the data falls below this value.
5.  **Maximum:** The largest value in the dataset (excluding outliers).

**Components of the Boxplot:**
*   The **box** itself represents the interquartile range (IQR), spanning from Q1 to Q3. The length of the box indicates the spread of the middle 50% of the data.
*   The **line inside the box** indicates the median.
*   The **whiskers** extend from the box to the minimum and maximum values within 1.5 times the IQR from Q1 and Q3, respectively. Data points outside these whiskers are considered potential outliers.
*   **Dots or asterisks** beyond the whiskers represent the identified **outliers**.

**Result Interpretation for the given data:**

*   **Q1 (25th percentile):** `18.75`
*   **Q3 (75th percentile):** `23.75`
*   **IQR (Interquartile Range):** `5.0` (23.75 - 18.75)
*   **Lower Bound for Outliers:** `11.25` (18.75 - 1.5 * 5.0)
*   **Upper Bound for Outliers:** `31.25` (23.75 + 1.5 * 5.0)

Based on these calculations, any data point less than `11.25` or greater than `31.25` is considered an outlier.

*   **Identified Outliers:** `[35]`

The boxplot would show the main cluster of data, with the value `35` plotted as a distinct point beyond the upper whisker, indicating it is an outlier.

# **Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.  Explain how you would use covariance and correlation to explore this relationship.  Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000] .**

### Exploring the Relationship between Advertising Spend and Daily Sales using Covariance and Correlation

As a data analyst, to understand the relationship between advertising spend and daily sales for the marketing team, I would use **covariance** and **correlation**.

#### 1. Covariance

**How it works:**
Covariance measures the directional relationship between two variables. In this case, it would tell us if advertising spend and daily sales tend to move in the same direction.

*   A **positive covariance** would suggest that as advertising spend increases, daily sales tend to increase, and vice-versa.
*   A **negative covariance** would suggest that as advertising spend increases, daily sales tend to decrease.
*   A covariance close to **zero** would suggest no linear relationship between the two variables.

**Interpretation for the marketing team:**
If we compute a positive covariance between advertising spend and daily sales, it indicates a direct relationship  spending more on ads generally leads to more sales. However, the magnitude of the covariance value is hard to interpret on its own because it's affected by the scale of the variables. For instance, a high covariance could simply mean the numbers for sales are large, not necessarily a strong relationship.

#### 2. Correlation Coefficient

**How it works:**
While covariance tells us the direction, the **correlation coefficient** (specifically, Pearson's correlation coefficient) tells us both the direction *and* the strength of the *linear* relationship between advertising spend and daily sales. It's a standardized measure, ranging from -1 to +1.

*   A correlation coefficient of **+1** indicates a perfect positive linear relationship (as one variable increases, the other increases proportionally).
*   A correlation coefficient of **-1** indicates a perfect negative linear relationship (as one variable increases, the other decreases proportionally).
*   A correlation coefficient of **0** indicates no linear relationship.
*   Values between 0 and 1 (or -1 and 0) indicate weaker positive (or negative) linear relationships.

**Interpretation for the marketing team:**
*   **Strong Positive Correlation (e.g., 0.7 to 1.0):** This would strongly suggest that increased advertising spend is directly and significantly associated with higher daily sales. This is valuable information, as it would support investing more in advertising.
*   **Moderate Positive Correlation (e.g., 0.3 to 0.69):** This indicates that advertising spend has some positive impact on sales, but other factors might also be at play, or the relationship isn't as strong.
*   **Weak or No Correlation (e.g., -0.29 to 0.29):** If the correlation is close to zero, it means there's little to no linear relationship. The marketing team would need to investigate other strategies or variables affecting sales.
*   **Negative Correlation (e.g., -0.3 to -1.0):** This would be highly unusual and concerning in this context, suggesting that increasing ad spend might actually lead to *decreased* sales. This would require immediate and thorough investigation.

**In summary,** I would first calculate the covariance to see the direction. Then, I would calculate the correlation coefficient to understand the strength of this relationship, which provides a more interpretable measure for decision-making. A high positive correlation would justify continued or increased investment in advertising, while a low or negative correlation would signal a need to re-evaluate the marketing strategy.

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to numpy arrays
array_advertising_spend = np.array(advertising_spend)
array_daily_sales = np.array(daily_sales)

# Compute the correlation coefficient
# np.corrcoef returns a correlation matrix.
# The correlation coefficient between advertising_spend and daily_sales is at index [0, 1] or [1, 0].
correlation_matrix = np.corrcoef(array_advertising_spend, array_daily_sales)
correlation_coefficient = correlation_matrix[0, 1]

print(f"Advertising Spend: {advertising_spend}")
print(f"Daily Sales: {daily_sales}")
print(f"Correlation Coefficient: {correlation_coefficient}")
```

It's a good complement to the histogram for quickly spotting spread and outliers.

By using these statistics and visualizations, the marketing team can gain a comprehensive understanding of current customer satisfaction, identify areas of concern, and make data-driven decisions before the new product launch.

```python
import matplotlib.pyplot as plt
import numpy as np
import collections

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate summary statistics
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)

# Calculate mode(s)
counts = collections.Counter(survey_scores)
max_count = max(counts.values())
mode_scores = [key for key, value in counts.items() if value == max_count]

std_dev = np.std(survey_scores)

print(f"Survey Scores: {survey_scores}")
print(f"Mean Score: {mean_score:.2f}")
print(f"Median Score: {median_score}")
print(f"Mode(s): {sorted(mode_scores)}")
print(f"Standard Deviation: {std_dev:.2f}")

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(survey_scores, bins=np.arange(3.5, 11.5, 1), edgecolor='black', alpha=0.7)
# np.arange(3.5, 11.5, 1) creates bins centered around integers for scores 4-10

plt.title('Distribution of Customer Satisfaction Survey Scores')
plt.xlabel('Satisfaction Score (1-10)')
plt.ylabel('Frequency')
plt.xticks(np.arange(4, 11, 1)) # Set x-axis ticks to actual score values
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
